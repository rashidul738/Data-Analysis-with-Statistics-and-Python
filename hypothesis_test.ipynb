```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
ed = pd.read_csv('education_districtwise.csv')
ed = ed.dropna()
```

```python
ed.head()
```

```python
state21 = ed[ed['STATNAME'] == 'STATE21']
```

```python
state28 = ed[ed['STATNAME'] == 'STATE28']
```

```python
sample_state21 = state21.sample(n = 20, replace=True, random_state = 13490)
sample_state21
```

```python
sample_state28 = state28.sample(n = 20, replace=True, random_state = 39103)
sample_state28
```

```python
sample_state21['OVERALL_LI'].mean()
```

```python
sample_state28['OVERALL_LI'].mean()
```

```python
stats.ttest_ind(a = sample_state21['OVERALL_LI'], b = sample_state28['OVERALL_LI'], equal_var = False)
```

```python
aqi = pd.read_csv('c4_epa_air_quality.csv')
aqi = aqi.dropna()
aqi.head()
```

```python
print('use head to see the first 5 rows of the dataframe')
```

```python
aqi.describe()
```

```python
aqi['state_name'].value_counts()
```

```python
ca_la = aqi[aqi['county_name'] == 'Los Angeles']
ca_other = aqi[(aqi['state_name'] == 'California') & (aqi['county_name']!= 'Los Angeles')]
```

```python
significance_level = 0.05
significance_level
```

```python
stats.ttest_ind(a=ca_la['aqi'], b=ca_other['aqi'], equal_var=False)
```

```python

ny = aqi[aqi['state_name']=='New York']
ohio = aqi[aqi['state_name']=='Ohio']
```

```python
tstat, pvalue = stats.ttest_ind(a=ny['aqi'], b=ohio['aqi'], alternative='less', equal_var=False)
print(tstat)
print(pvalue)
```

```python
michigan = aqi[aqi['state_name']=='Michigan']
```

```python
tstat, pvalue = stats.ttest_1samp(michigan['aqi'], 10, alternative='greater')
print(tstat)
print(pvalue)
```
